## Data Filtering

```python
zf.drop_duplicates(inplace=True) # drop duplicates
```

```python
zf[zf['property_type'] == 'House']
```

```python
zf[zf['price'] < 10000000] # filter data
```

```python
zf.rename(columns={'page_url': 'URL' , 'province_name': 'province'}, inplace=True) # rename column
```

## Data Cleaning

```python
zf.isnull().sum() # missing values
```

```python
zf[zf['agency'].isnull()] # missing values
```

```python
zf['agency'].fillna(value="Not Mentioned",inplace=True) # fill missing values
```

```python
zf['agent'].fillna(value="Not Mentioned",inplace=True) # fill missing values
```

```python
zf.isnull().sum() # missing values
```

```python
zf. drop(columns=["Unnamed: 0" , "purpose"] , inplace=True) # drop columns
```

```python
zf.groupby('province').describe() # group by
```
